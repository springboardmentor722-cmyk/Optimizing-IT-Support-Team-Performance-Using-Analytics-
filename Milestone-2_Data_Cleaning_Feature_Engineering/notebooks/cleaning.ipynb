```python
import pandas as pd
df = pd.read_excel("customer_support_tickets.xlsx", index_col=None)
data = df.drop(["Customer Email", "Unnamed: 9", "Resolution", "First Response Time", "Customer Satisfaction Rating"],axis=1)

data["Date of Purchase"] = pd.to_datetime(data["Date of Purchase"])

data["Time to Resolution"] = pd.to_datetime(data["Time to Resolution"])
data
```

```python
data['Time to Resolution'] = pd.to_datetime(data['Time to Resolution'])

data['Resolution_Time_Hours'] = (
    data['Time to Resolution'].dt.hour +
    data['Time to Resolution'].dt.minute / 60 +
    data['Time to Resolution'].dt.second / 3600
).round(2)

data
```

```python

valid_times = data.loc[data['Resolution_Time_Hours'] > 1, 'Resolution_Time_Hours']

min_time = valid_times.min()

data['Response_Efficiency_%'] = (
    (min_time / data['Resolution_Time_Hours']) * 100
).round(2)

data
```

```python
priority_mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Critical': 4
}

data['Priority_Score'] = data['Ticket Priority'].map(priority_mapping)

data
```

```python
priority_distribution = (
    df['Ticket Priority']
    .value_counts()
    .to_frame(name='Ticket_Count')
    .reset_index()
)

priority_distribution['Percentage (%)'] = (
    priority_distribution['Ticket_Count'] / len(df) * 100
).round(2)

priority_distribution
```

```python
import pandas as pd

def categorize_time(hours):
    if pd.isna(hours):
        return "Pending"
    elif hours <= 4:
        return "Fast"
    elif hours <= 12:
        return "Medium"
    else:
        return "Slow"

data['Resolution_Category'] = data['Resolution_Time_Hours'].apply(categorize_time)

data
```

```python
data.to_excel("new_dataset.xlsx", index=False)
```
