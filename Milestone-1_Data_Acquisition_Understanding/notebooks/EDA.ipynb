```python
import pandas as pd
```

```python
df = pd.read_excel("new_dataset.xlsx")
df
```

```python
df['Ticket Type'].value_counts()
```

```python
import matplotlib.pyplot as plt

tt = df['Ticket Type'].value_counts().plot(kind='barh')
plt.title("Ticket Type Distribution")
for container in tt.containers:
    tt.bar_label(container, label_type='center')
plt.show()
```

```python
tp = df['Ticket Priority'].value_counts().plot(kind='barh')
plt.title("Ticket Volume by Priority")
for container in tp.containers:
    tp.bar_label(container, label_type='center')
plt.show()
```

```python
ax = df['Ticket Channel'].value_counts().plot(kind='barh', color='wheat')

plt.title("Ticket Volume by Channel")
plt.xlabel("Channel")
plt.ylabel("Number of Tickets")

for container in ax.containers:
    ax.bar_label(container, label_type='center')

plt.show()
```

```python
issue_percentage = (
    df['Ticket Subject'].value_counts(normalize=True) * 100
).round(2)

print(issue_percentage.head(10))
```
