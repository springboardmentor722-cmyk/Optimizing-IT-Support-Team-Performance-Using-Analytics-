EXPLORATORY VISUALIZATION

Ticket Type Distribution → Pie Chart

```python
type_counts = df['Ticket Type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%')
plt.title("Ticket Type Distribution")
plt.show()
```

Top Categories (Ticket Subject) → Horizontal Bar Chart

```python
top_subjects = df['Ticket Subject'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.barh(top_subjects.index, top_subjects.values)
plt.title("Top 10 Ticket Categories")
plt.xlabel("Number of Tickets")
plt.ylabel("Category")
plt.gca().invert_yaxis()
plt.show()
```

Ticket Priority → Donut Chart 

```python
priority_counts = df['Ticket Priority'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(priority_counts, labels=priority_counts.index, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Ticket Priority Distribution")
plt.show()
```

Ticket Channel → Bar Chart

```python
channel_counts = df['Ticket Channel'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(channel_counts.index, channel_counts.values)
plt.title("Tickets by Channel")
plt.xticks(rotation=30)
plt.show()
```

Priority vs Type → Stacked Bar Chart

```python
cross_tab = pd.crosstab(df['Ticket Type'], df['Ticket Priority'])

cross_tab.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Ticket Type vs Priority (Stacked)")
plt.ylabel("Count")
plt.show()
```

PERFORMANCE & GAP ANALYSIS

Resolution Time Distribution → Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(df['Time to Resolution'].dropna(), bins=20)
plt.title("Distribution of Resolution Time")
plt.xlabel("Time to Resolution")
plt.ylabel("Frequency")
plt.show()
```

Resolution Time by Priority → Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Ticket Priority', y='Time to Resolution', data=df)
plt.title("Resolution Time by Priority")
plt.show()
```

Response Time vs Resolution Time → Scatter Plot

```python
plt.figure(figsize=(8,5))
plt.scatter(df['First Response Time'], df['Time to Resolution'])
plt.title("Response Time vs Resolution Time")
plt.xlabel("First Response Time")
plt.ylabel("Time to Resolution")
plt.show()
```

Correlation Heatmap

```python
# Convert datetime to numeric
df['First Response Time_num'] = df['First Response Time'].astype('int64')
df['Time to Resolution_num'] = df['Time to Resolution'].astype('int64')

corr = df[['First Response Time_num', 'Time to Resolution_num']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Average Resolution Time by Channel → Bar Plot

```python
avg_resolution = df.groupby('Ticket Channel')['Time to Resolution'].mean()

plt.figure(figsize=(6,4))
avg_resolution.plot(kind='bar')
plt.title("Average Resolution Time by Channel")
plt.ylabel("Time")
plt.show()
```

PERFORMANCE TREND ANALYSIS

Average Resolution Time by Priority

```python
df['Time to Resolution'] = pd.to_datetime(df['Time to Resolution'])
df['First Response Time'] = pd.to_datetime(df['First Response Time'])
```

Calculate duration

```python
df['Resolution_Duration'] = (
    df['Time to Resolution'] - df['First Response Time']
).dt.total_seconds() / 3600  # in hours
```

Average by Priority

```python
avg_priority = df.groupby('Ticket Priority')['Resolution_Duration'].mean()
avg_priority
```

Visualization → Bar Chart

```python
avg_priority.plot(kind='bar', figsize=(6,4))
plt.title("Average Resolution Time by Priority")
plt.ylabel("Hours")
plt.show()
```

Average Resolution Time by Ticket Type

avg_type = df.groupby('Ticket Type')['Resolution_Duration'].mean()

plt.ylabel("Hours")
plt.show()

Who Handles Tickets Fastest? (Channel = Team)

```python
fastest_channel = df.groupby('Ticket Channel')['Resolution_Duration'].mean().sort_values()

fastest_channel
```
